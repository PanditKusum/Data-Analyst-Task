```python
# task 2  find and Replace the missing values by running average

import pandas as pd
from IPython.display import display

# Load Project CSV
project_df = pd.read_csv("Project.csv")

# Initialize empty list to store running average
running_avg = []

# Initialize sum and count
total = 0
count = 0

# use loop in a row
for i in range(len(project_df)):
    
    # If value is not null
    if not pd.isna(project_df.loc[i, "Cost"]):
        total += project_df.loc[i, "Cost"]  # Add cost
        count += 1  # Increase count
        running_avg.append(project_df.loc[i, "Cost"])  # Keep original value
    
    else:
        avg = total / count  # Calculate running average
        running_avg.append(avg)  # Replace missing value

# Replace Cost column
project_df["Cost"] = running_avg

# Save updated file

project_df.set_index("ID", inplace=True)
project_df.to_csv("Project.csv", index=True)

# Display output
print ("project_df:")
display (project_df)
```

```python
# task 3 split the column into two parts and remove the older name column

# import libraries
import pandas as pd 

# load the employee csv files
employee_df =pd.read_csv("employee.csv").set_index("ID")

# split firstname and last name
employee_df [["FirstName","LastName"]]=employee_df["Name"].str.split(" ",expand=True)

# Drop original Name column
employee_df.drop("Name", axis=1, inplace=True)

employee_df = employee_df[["FirstName", "LastName", "Gender","City", "Age"]]

# Save updated CSV
employee_df.to_csv("employee1.csv", index=False)

# Display output
print ("new employeedata:")
display (employee_df)
```

```python
# Task 4  Join all three dataframes in one single dataframe. Name it “Final

# import python libraries 
import pandas as pd
from IPython.display import display

# load all three  csv file 
project_df=pd.read_csv("Project.csv").set_index("ID")
employee_df=pd.read_csv("employee.csv").set_index("ID")
seniority_df=pd.read_csv ("seniority.csv").set_index("ID")

# Merge Project and Employee on 'ID'
df1 = pd.merge(project_df, employee_df, on="ID")   # Inner join

# Merge the above result with Seniority dataframe
Final = pd.merge(df1,seniority_df, on="ID")    # Final merged dataframe

# Display Final dataframe
print("Final DataFrame:")
display(Final)
```

```python
# Task 5 Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
#cost only to employees who have finished the projects

# import python libraries
import pandas as pd
from IPython.display import display
import numpy as np

# Create Bonus column based on condition
Final["Bonus"] = np.where(
    (Final["Status"] == "Finished") & (Final["Cost"].notna()),  # Condition
    Final["Cost"] * 0.05,  # 5% of cost
    0  # Otherwise 0
)

# Display updated dataframe
print("Final DataFrame with Bonus column:")
print(Final)
```

```python
#task 6
#Demote the designation level by 1, whose projects have status “fail”. Delete the
#employees record whose designation level is above 4

# Import required libraries
import pandas as pd

# Load CSV files
project = pd.read_csv("Project.csv")       # Load Project data
employee = pd.read_csv("Employee.csv")     # Load Employee data
seniority = pd.read_csv("Seniority.csv")   # Load Seniority data

# Create Final DataFrame (Task 4)
df1 = pd.merge(project, employee, on="ID")        # Merge Project + Employee
Final = pd.merge(df1, seniority, on="ID")         # Merge with Seniority

# Task 6: Demote designation level where status is 'Failed'
Final.loc[Final["Status"] == "Failed", "Designation Level"] -= 1

# Delete employees whose designation level is above 4
Final = Final[Final["Designation Level"] <= 4]

# Display result
print("newDataframe:")
print(Final)

# Save to new CSV file
Final.to_csv("Final_Task6.csv", index=False)
```

```python
# Task 7 Add Mr./Mrs. and Drop Gender
# Import python libraries 
import numpy as np 
# Add prefix based on gender
Final["Name"] = np.where(
    Final["Gender"] == "M",        
    "Mr. " + Final["Name"],  
    "Mrs. " + Final["Name"]   
)
 #Drop Gender column
Final.drop("Gender", axis=1, inplace=True)

# Display updated dataframe
print("updatedDataframe1:")
print(Final)
```

```python
# Task 8 Promote designation level by 1 for the employees whose age is more than 29 years using

# import python libraries
import pandas as pd

Final.loc[Final["Age"] > 29, "Designation Level"] += 1

# Display updated dataframe
print("\nAfter promotion based on age:")
print(Final)
```

```python
# Task 9 : Total Project Cost per Employee
#import python libraries 
import pandas as pd
#Group by ID and Name, sum cost
TotalProjCost = Final.groupby(["ID", "Name"])["Cost"].sum().reset_index()

# Rename column
TotalProjCost.rename(columns={"Cost": "Total Cost"}, inplace=True)

# Display result
print("\nTotal Project Cost DataFrame:")
print(TotalProjCost)

```
